```python
import csv

coordSet = set()
monthDict = {}

with open('17778.csv') as fp:
    colors = ['red','white','blue', 'yellow', 'cyan', 'magenta
    savedRows = []
    julyRows = []
    augustRows = []
    septemberRows = []
    reader = csv.DictReader(fp)
    outputFileNames = ['julySandshark.csv', 'augustSandshark.csv','septemberSandshark.csv']
    for row in reader:
        #print (row)
        if "Transmitter" in row:
            del row['Transmitter']
        if (row['Latitude'],row['Longitude']) not in coordSet:
            if row['Datetime_UTC'].split('/')[0]==7:
                row['LineColor'] = 'red'
                julyRows.append(row)
            elif row['Datetime_UTC'].split('/')[0]==8:
                row['LineColor'] = 'white'
                augustRows.append(row)
            elif row['Datetime_UTC'].split('/')[0]==9:
                row['LineColor'] = 'blue'
                septemberRows.append(row)
            savedRows.append(row)
            coordSet.add((row['Latitude'],row['Longitude']))
    print(coordSet)
    print(savedRows)
    with open('uniques.csv', 'w', newline='') as csvfile:
        fieldnames = savedRows.keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for row in savedRows:
            writer.writerow(row)
    
```
